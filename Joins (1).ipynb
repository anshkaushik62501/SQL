**8.Question 8. Show all customers along with order and payment amounts in one table.
**

**ans-**

SELECT *

FROM customers c

LEFT JOIN orders o

ON c.customer_id = o.customer_id

LEFT JOIN payments p

ON c.customer_id = p.customer_id;
