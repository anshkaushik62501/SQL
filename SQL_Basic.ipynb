**Q10. What Are Constraints in SQL and Why Are They Used?**

**Ans-**


Constraints in SQL are rules applied to table columns to restrict the type of data that can be inserted into a table. They help in maintaining the accuracy, consistency, and integrity of data in the database.

Constraints are used to prevent invalid data entry, avoid duplicate records, and enforce relationships between tables. By using constraints, databases remain reliable and errors are minimized.

Common SQL Constraints include:

* PRIMARY KEY  uniquely identifies each record

* FOREIGN KEY  maintains relationship between tables

* NOT NULL  prevents NULL values

* UNIQUE  avoids duplicate values

* CHECK  validates data based on a condition

* DEFAULT  assigns a default value

* AUTO_INCREMENT (MySQL)

Automatically generates numeric values
